#Visualizations to uncover patterns and relationships within the data.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization 1: Distribution of house prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True, color='blue')
plt.title('Distribution of House Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()

```

```python
# Ensure only numerical columns are included in the correlation
numerical_columns = df.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = numerical_columns.corr()

# Visualization: Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()

```

```python
# Visualization 3: Scatter plot of living area (sqft) vs. price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=df, alpha=0.7, color='green')
plt.title('Living Area (sqft) vs. Price')
plt.xlabel('Living Area (sqft)')
plt.ylabel('Price ($)')
plt.show()
```
